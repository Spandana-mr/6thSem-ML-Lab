With Library

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

data = pd.read_csv("iris (1).csv")

X = data.iloc[:, :-1]
y = data.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)

model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train,y_train)

y_pred = model.predict(X_test)

print("Accuracy:",accuracy_score(y_test,y_pred))
print("Confusion Matrix\n",confusion_matrix(y_test,y_pred))
print("Classification Report\n",classification_report(y_test,y_pred))




data = pd.read_csv("diabetes.csv")

X = data.iloc[:, :-1]
y = data.iloc[:, -1]

scaler = StandardScaler()
X = scaler.fit_transform(X)

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train,y_train)

y_pred = model.predict(X_test)

print("Accuracy:",accuracy_score(y_test,y_pred))
print("Confusion Matrix\n",confusion_matrix(y_test,y_pred))





data = pd.read_csv("heart.csv")

X = data.iloc[:, :-1]
y = data.iloc[:, -1]

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

best_k = 1
best_acc = 0

for k in range(1,20):
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train,y_train)
    pred = model.predict(X_test)
    acc = accuracy_score(y_test,pred)

    if acc > best_acc:
        best_acc = acc
        best_k = k

print("Best K:",best_k)

model = KNeighborsClassifier(n_neighbors=best_k)
model.fit(X_train,y_train)

y_pred = model.predict(X_test)

print("Accuracy:",accuracy_score(y_test,y_pred))
print("Confusion Matrix\n",confusion_matrix(y_test,y_pred))
print("Classification Report\n",classification_report(y_test,y_pred))
```

```python
import pandas as pd
import math
from collections import Counter

data = pd.read_csv("iris (1).csv")

X = data.iloc[:,:-1].values
y = data.iloc[:,-1].values

split = int(0.8*len(X))

X_train = X[:split]
X_test = X[split:]
y_train = y[:split]
y_test = y[split:]

def distance(a,b):
    s = 0
    for i in range(len(a)):
        s += (a[i]-b[i])**2
    return math.sqrt(s)

def predict(test,k):
    d = []

    for i in range(len(X_train)):
        dist = distance(test,X_train[i])
        d.append((dist,y_train[i]))

    d.sort()

    neighbors = []
    for i in range(k):
        neighbors.append(d[i][1])

    return Counter(neighbors).most_common(1)[0][0]

k = 5
pred = []

for t in X_test:
    pred.append(predict(t,k))

correct = 0
for i in range(len(y_test)):
    if pred[i] == y_test[i]:
        correct += 1

print("Accuracy:",correct/len(y_test))






data = pd.read_csv("diabetes.csv")

X = data.iloc[:,:-1].values
y = data.iloc[:,-1].values

split = int(0.8*len(X))

X_train = X[:split]
X_test = X[split:]
y_train = y[:split]
y_test = y[split:]

def distance(a,b):
    s=0
    for i in range(len(a)):
        s += (a[i]-b[i])**2
    return math.sqrt(s)

def predict(test,k):
    dist=[]

    for i in range(len(X_train)):
        d = distance(test,X_train[i])
        dist.append((d,y_train[i]))

    dist.sort()

    neighbors=[]
    for i in range(k):
        neighbors.append(dist[i][1])

    return Counter(neighbors).most_common(1)[0][0]

k=5
pred=[]

for t in X_test:
    pred.append(predict(t,k))

correct=0
for i in range(len(y_test)):
    if pred[i]==y_test[i]:
        correct+=1

print("Accuracy:",correct/len(y_test))







data = pd.read_csv("heart.csv")

X = data.iloc[:,:-1].values
y = data.iloc[:,-1].values

split = int(0.8*len(X))

X_train = X[:split]
X_test = X[split:]
y_train = y[:split]
y_test = y[split:]

def distance(a,b):
    s=0
    for i in range(len(a)):
        s += (a[i]-b[i])**2
    return math.sqrt(s)

def predict(test,k):
    dist=[]
    for i in range(len(X_train)):
        d = distance(test,X_train[i])
        dist.append((d,y_train[i]))

    dist.sort()

    neighbors=[]
    for i in range(k):
        neighbors.append(dist[i][1])

    return Counter(neighbors).most_common(1)[0][0]

best_k=1
best_acc=0

for k in range(1,10):
    pred=[]

    for t in X_test:
        pred.append(predict(t,k))

    correct=0
    for i in range(len(y_test)):
        if pred[i]==y_test[i]:
            correct+=1

    acc = correct/len(y_test)

    if acc > best_acc:
        best_acc = acc
        best_k = k

print("Best K:",best_k)
```
