```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from scipy import stats

diabetes = pd.read_csv("diabeteees.csv")

cols_with_zero = ['BMI']
diabetes[cols_with_zero] = diabetes[cols_with_zero].replace(0, np.nan)

diabetes.fillna(diabetes.mean(numeric_only=True), inplace=True)

numeric_cols_for_zscore = diabetes.select_dtypes(include=np.number)
z_scores = np.abs(stats.zscore(numeric_cols_for_zscore))

mask = (z_scores < 3).all(axis=1)
diabetes = diabetes[mask]

numeric_cols_for_scaling = diabetes.select_dtypes(include=np.number)

minmax = MinMaxScaler()

diabetes_minmax = pd.DataFrame(minmax.fit_transform(numeric_cols_for_scaling),
                                columns=numeric_cols_for_scaling.columns,
                                index=numeric_cols_for_scaling.index)

standard = StandardScaler()

diabetes_standard = pd.DataFrame(standard.fit_transform(numeric_cols_for_scaling),
                                  columns=numeric_cols_for_scaling.columns,
                                  index=numeric_cols_for_scaling.index)

print("Diabetes preprocessing completed")

```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder
from scipy import stats

adult = pd.read_csv("adult.csv")

adult.replace("?", np.nan, inplace=True)

for col in adult.columns:
    if adult[col].isnull().sum() > 0:
        adult[col] = adult[col].fillna(adult[col].mode()[0])

categorical_cols = adult.select_dtypes(include=['object']).columns

le = LabelEncoder()
for col in categorical_cols:
    adult[col] = le.fit_transform(adult[col])

z_scores = np.abs(stats.zscore(adult))
adult = adult[(z_scores < 3).all(axis=1)]

minmax = MinMaxScaler()
adult_minmax = pd.DataFrame(
    minmax.fit_transform(adult),
    columns=adult.columns
)

standard = StandardScaler()
adult_standard = pd.DataFrame(
    standard.fit_transform(adult),
    columns=adult.columns
)

print("Adult income preprocessing completed")
```
